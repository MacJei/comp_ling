```
from yargy import Parser, rule, or_
from yargy.predicates import in_, in_caseless
from yargy.tokenizer import MorphTokenizer
from yargy.pipelines import morph_pipeline, caseless_pipeline
from yargy.interpretation import fact
from IPython.display import display
```

```
!wget https://raw.githubusercontent.com/mannefedov/compling_nlp_hse_course/master/data/pristavki.csv
```

```
import pandas as pd
pd.set_option('display.max_colwidth', -1)
```

```
data = pd.read_csv('pristavki.csv.3', header=None, names=['text'])
```

```
data.head(10)
```

```

Game = fact(
    'Game',
    ['name', 'model']
)
```

```
AssassinsCreed = rule(
    morph_pipeline([ 'Assassins Creed', 'Assassin\'s creed', 'асассин',]).interpretation(Game.name),
    morph_pipeline(['4', '4 Черный флаг', 'IV', 'Rogue']).interpretation(Game.model)    
    )
```

```
FIFA = rule(
    morph_pipeline(['фифа', 'Фифв', "FIFA"]).interpretation(Game.name),
    morph_pipeline(['15,16', '13']).interpretation(Game.version))
```

```
MortalCombat = rule(
    morph_pipeline([ 'Mortal combat', 'мортал комбат']).interpretation(Game.version),
    morph_pipeline(['vs D.C.', '9', 'Х', 'XL']).interpretation(Game.version)    
    )
```

```
Watchdogs = rule(
    morph_pipeline([ 'вотч догс', 'Watch Dogs']).interpretation(Game.name),
    morph_pipeline(['2']).interpretation(Game.version)    
    )
```

```
GTA = rule(
    morph_pipeline(['гта', 'GTA']).interpretation(Game.name),
    morph_pipeline(['5', '4']).interpretation(Game.version))
```

```
Igra = or_(FIFA, GTA, Watchdogs, MortalCombat, AssassinsCreed).interpretation(Game)
parser = Parser(Igra)
```

```
matches = []

for sent in data.text[:9000]:
    for match in parser.findall(sent):
        matches.append(match.fact)
```

```
for m in matches:
    print(m.name, m.model)
```

Данный метод довольно эффективный для неструктурированных данных, так как в процессе извлечения именованных сущностей не извлекается ничего лишнего, то есть качество извлеченных именованных сущностей довольно высокое. К плюсам такого подхода также можно отнести то, что сущности находятся в независимости от регистра. Однако стоит заметить, что процесс извлечения названий игр довольно долгий, зачастую в самих данных сложно определить, что является названием, а что - версией(изданием). Всё это сильно затрудняет процесс.
